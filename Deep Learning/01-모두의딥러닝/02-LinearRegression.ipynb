# Linear Regression

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]
```

## 모델 구축

```python
# 64개의 유닛을 가진 완전연결 층을 모델에 추가
tf.model = tf.keras.Sequential()
# 또하나의 층을 추가한다는 의미, 객체 생성 
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))
```

* Dense Layer: Input과 Output을 모두 연결해주는 층
* units: output shape
* input_dim: output shape

## 모델 구현

```python
sgd = tf.keras.optimizers.SGD(lr=0.1)
tf.model.compile(loss='mse', optimizer=sgd)
```

**Optimizer**
: 학습에 최적화 알고리즘을 결정

```python
tf.model.summary()
```

## 모델 학습

```python
models = tf.model.fit(x_train,y_train,epochs=200)
```

* epoch: 학습 반복 횟수
* loss가 0에 가까울 수록 좋은 모델 
* accuracy는 1에 가까울 수록 좋은 모델

```python
# returns predicted value
y_predict = tf.model.predict(np.array([5,4]))
y_predict
```
